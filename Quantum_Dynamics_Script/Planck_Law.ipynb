```python
import numpy as np
import matplotlib.pylab as plt

plt.rc('text', usetex=True)
```

```python
pi = 3.14159
h = 6.62607*10**(-34)
c = 299792458
k = 1.380649*10**(-23)

def planck_law(T, lamb):
    return 2*pi*h*c**2/(lamb**5)*1/(np.exp(h*c/(lamb*k*T))-1)
```

```python
fig = plt.figure(figsize =(13, 11))
ax = fig.add_subplot(1,1,1)
for n in [100, 200, 250, 300, 350, 400]:
    x = np.linspace(1*10**(-15), 31.0*10**(-6), 10000.0)
    y = planck_law(n, x)
    ax.plot(x, y, label=r'$%d\;\mathrm{K}$'%n)

ax.set_xlim(0.0, 31*10**(-6))
ax.set_xticks(np.linspace(0.0, 0.00003, 7, endpoint=True))
ax.set_xticklabels([r'$%d$'%(0 + i*5) for i in range(7)], size=20)

ax.set_ylim(-0.05*10**8, 1.42*10**8)
ax.set_yticks(np.linspace(0.0, 1.4*10**8, 8, endpoint=True))
ax.set_yticklabels([r'$%d$'%(0 + i*2) for i in range(8)], size=20)


plt.legend(loc=(0.65, 0.50), fontsize=20).get_frame().set_edgecolor('k')


ax.set_xlabel(r'$\mathrm{Wavelength}\,\mathrm{[\mu m]}$', fontsize=23, labelpad=12)
ax.set_ylabel(r'$\mathrm{Spectral}\;\mathrm{Radiance}\;\Bigg[10^7\;\mathrm{W}/\mathrm{m}^3\Bigg]$', fontsize=23, labelpad=25)





#plt.show()
fig.savefig("2-Planck-Strahlungsgesetz.pdf", bbox_inches="tight")
```
